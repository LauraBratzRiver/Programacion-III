```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
longitudes = pd.read_csv("/content/longitud.csv", sep=",")
masas = pd.read_csv("/content/masa.csv", sep=",")
almacenamiento = pd.read_csv("/content/almacenamiento.csv", sep=",")
tiempo = pd.read_csv("/content/tiempo.csv", sep=",")
area = pd.read_csv("/content/area.csv", sep=",")
meses_acumulados = pd.read_csv("/content/meses.csv", sep=",")
print(longitudes)
print(masas)


sns.scatterplot(longitudes["metros"], longitudes["centimetros"])
plt.show()
```

```python
metros = longitudes["metros"]
centimetros = longitudes["centimetros"]
libras = masas["libras"]
gramos = masas["gramos"]
megabytes = almacenamiento["megabytes"]
gigabytes = almacenamiento["gigabytes"]
segundos = tiempo["segundos"]
minutos = tiempo["minutos"]
metros_cuadrados = area["metros_cuadrados"]
centimetros_cuadrados = area["centimetros_cuadrados"]
meses = meses_acumulados["meses"]
acumulacion = meses_acumulados["acumulacion"]
```

```python
modeloLongitud = tf.keras.Sequential()
modeloLongitud.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
modeloMasa = tf.keras.Sequential()
modeloMasa.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
modeloDatos = tf.keras.Sequential()
modeloDatos.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
modeloTiempo = tf.keras.Sequential()
modeloTiempo.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
modeloArea = tf.keras.Sequential()
modeloArea.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
modeloMeses = tf.keras.Sequential()
modeloMeses.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
```

```python
modeloLongitud.compile(optimizer=tf.keras.optimizers.Adam(1), loss="mean_squared_error")
modeloMasa.compile(optimizer=tf.keras.optimizers.Adam(1), loss="mean_squared_error")
modeloDatos.compile(optimizer=tf.keras.optimizers.Adam(1), loss="mean_squared_error")
modeloTiempo.compile(optimizer=tf.keras.optimizers.Adam(1), loss="mean_squared_error")
modeloArea.compile(optimizer=tf.keras.optimizers.Adam(1), loss="mean_squared_error")
modeloMeses.compile(optimizer=tf.keras.optimizers.Adam(1), loss="mean_squared_error")
```

```python
historialLongitud = modeloLongitud.fit(metros, centimetros, epochs=100)
historialMasa = modeloMasa.fit(libras, gramos, epochs=100)
historialDatos = modeloDatos.fit(megabytes, gigabytes, epochs=100)
historialTiempo = modeloTiempo.fit(segundos, minutos, epochs=100)
historialArea = modeloArea.fit(metros_cuadrados, centimetros_cuadrados, epochs=100)
historialMeses = modeloMeses.fit(meses, acumulacion, epochs=100)
```

```python
f = modeloLongitud.predict([2])
g = modeloMasa.predict([45])
d = modeloDatos.predict([10])
t = modeloTiempo.predict([7])
a = modeloArea.predict([9])
m = modeloMeses.predict([9])
print(f)
print(g)
print(d)
print(t)
print(a)
print(m)
```

```python
plt.plot(historialLongitud.history["loss"])
plt.title("Progreso del aprendizaje")
plt.xlabel("vuelta")
plt.ylabel("error de aprendizaje")
plt.legend("Aprendizaje Automatico")
plt.show()
```
