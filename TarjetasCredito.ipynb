```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
archivo = pd.read_csv("/content/trx_tarjetas_credito.csv", sep=",")

getGender = pd.crosstab(index=archivo["Gender"], columns="Gender")
print(getGender)

pieGender = getGender.loc[getGender.index]["Gender"]
plt.pie(pieGender, autopct="%1.1f%%", shadow=True, startangle=90, explode=(0.1, 0))
plt.legend(getGender.index,
           title="Generos",
           loc="center left",
           bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("Uso de tarjetas por genero")

plt.show()
getGender.to_excel("Credit_card_gender.xlsx")
```

```python
getCity = archivo.groupby(["City", "Card Type"]).sum()
print(getCity)

sepCity = getCity[:10]
pieCity = sepCity.loc[sepCity.index]
plt.pie(pieCity, autopct="%1.1f%%", shadow=True, startangle=90, explode=(0.07, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07))

plt.legend(getCity.index,
           title="Ciudades y tipos de tarjeta",
           loc="center left",
           bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("Tipos de tarjeta")
plt.show()
getCity.to_excel("Credit_card_city.xlsx")
```

```python
getUse = pd.crosstab(index=archivo["Exp Type"], columns="Use")
print(getUse)

pieUse = getUse.loc[getUse.index]["Use"]
plt.pie(pieUse, autopct="%1.1f%%", shadow=True, startangle=90, explode=(0.07, 0.07, 0.07, 0.07, 0.07, 0.07))

plt.legend(pieUse.index,
           title="Gastos",
           loc="center left",
           bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("Uso de tarjetas por gasto")
plt.show()

getUse.to_excel("Credit_card_use.xlsx")
```
